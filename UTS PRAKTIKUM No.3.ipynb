### Import Library

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

### Membaca Gambar

```python
img = cv2.imread('Nomor 3.jpg')
```

```python
img.shape
```

```python
[baris, kolom] = img.shape[:2]
```

### Convert BGR to RGB

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Convert to GrayScale

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

### Memperbaiki Gambar Backlight

```python
bright = cv2.convertScaleAbs(gray, alpha=1.5, beta=100)
```

```python
contrast = cv2.convertScaleAbs(gray, alpha=2.0, beta=75)
```

```python
bright_contrast = cv2.convertScaleAbs(gray, alpha=2.0, beta=150)
```

### Menampilkan Hasil

```python
titles = ['Gambar Asli', 'Gambar Gray', 'Gray yang Dicerahkan', 'Gray yang Diperkontras', 'Gambar yang Dipercerah dan Diperkontras']
images = [rgb, gray, bright, contrast, bright_contrast]

plt.figure(figsize=(14, 8))
for i in range(len(images)):
    plt.subplot(2, 3, i+1)
    if i == 0:
        plt.imshow(images[i])
    else:
        plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 8))
for i in range(len(images)):
    plt.subplot(2, 3, i+1)
    
    if i == 0:
        for j, color in enumerate(['red', 'green', 'blue']):
            plt.hist(images[i][:,:,j].ravel(), bins=256, color=color, alpha=0.5, label=color)
        plt.title("Histogram Gambar Asli (RGB)")
        plt.legend()
    else:
        plt.hist(images[i].ravel(), bins=256, color='gray', alpha=0.7)
        plt.title(f"Histogram {titles[i]}")
    
    plt.xlabel('Intensitas')
    plt.ylabel('Jumlah Piksel')
    plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
